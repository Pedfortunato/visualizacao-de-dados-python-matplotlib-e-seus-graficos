```python
# Criando um gráfico na mesma figura com subplot
# Bibliotecas necessarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variaveis e funções
X = np.linspace(0,5,11)  #variavel

Y = np.power(X-2,2.0)  #funções
Z = np.exp(X/2.0)
```

```python
# Plotando

plt.subplot(2,2,1)
plt.plot(X,Y)
plt.subplot(2,2,2)
plt.plot(X,Z)
plt.show()
```

```python
# Nomeando os eixos

plt.subplot(2,2,1)
plt.plot(X,Y)
plt.xlabel('Numeração')
plt.ylabel('Possibilidade')
plt.subplot(2,2,2)
plt.plot(X,Z)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Z')
plt.tight_layout()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(X,Y)
plt.xlabel('P')
plt.ylabel('B')
plt.subplot(1,2,2)
plt.plot(X,Z)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Z')
plt.tight_layout()
plt.show()
```

```python
# Usando notação de objeto

fig,ax = plt.subplots(nrows=1,ncols=8)
ax[0].plot(X,Y)
ax[0].set_xlabel('vcxl')
ax[0].set_ylabel('Eixo Y')
ax[1].plot(X,Z)
ax[1].set_xlabel('Eixo X')
ax[1].set_ylabel('Eixo Z')
plt.tight_layout()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1)
ax[0].plot(X,Y)
ax[0].set_xlabel('Eixo X')
ax[0].set_ylabel('Eixo Y')
ax[1].plot(X,Z)
ax[1].set_xlabel('Eixo X')
ax[1].set_ylabel('Eixo Z')
plt.tight_layout()
```
