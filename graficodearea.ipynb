```python
# Bibliotecas necessarias

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variaveis x e y

X = np.linspace(-5,5,101)
Y = (X-2.0)*(X+2.0)
```

```python
# Plotando o gráfico

plt.plot(X,Y)
plt.xlabel('Base', color = 'green')
plt.ylabel('Altura', color = 'red')
plt.fill_between(X, Y, color = 'black')
plt.show()
```
