```python
# Criando Gráfico de espelhamento
```

```python
#importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Criando variável X
X = np.linspace(0,5,11)
```

```python
#Criando funções Y(X) e Z(X)
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
```

```python
#Grafico de y em função de X
plt.scatter(X,Y)
plt.show()
```

```python
# Dando nomes aos eixos
plt.scatter(X,Y)
plt.xlabel('nome criativo')
plt.ylabel('nome mais criativo ainda')
plt.show()
```

```python
# Colocando titulo
plt.scatter(X,Y)
plt.xlabel('nome criativo')
plt.ylabel('nome mais criativo ainda')
plt.title('espelhamento')
plt.show()
```

```python
# Dois gráficos em uma unica imagem
plt.scatter(X,Y)
plt.scatter(X,Z)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
# Alterando Cores
plt.scatter(X,Y,color='black')
plt.scatter(X,Z,color='pink')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
# Determinando uma legenda aos eixos
plt.scatter(X,Y,color='black',label='Y(X)')
plt.scatter(X,Z,color='red',label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```
