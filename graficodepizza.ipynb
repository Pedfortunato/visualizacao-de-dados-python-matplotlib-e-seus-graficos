```python
# Precisaremos das seguintes bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Veremos o dominio dos fornecedores de material esportivo dos clubes da série A
```

```python
labels = ["Adidas", "Umbro", "Nike", "Puma" , "kappa","New","rebook","Marcas dos proprios clubes"]
```

```python
valores = [5,5,1,1,1,1,1,5]
```

```python
plt.pie(valores,labels=labels)
plt.show()
```

```python
# Agora com porcentagem
plt.pie(valores,labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
# Dando destaque a alguma categoria
explode = [0,0,0,0,0,0,0,0.1]
```

```python
plt.pie(valores,labels=labels, autopct='%1.1f%%',explode=explode)
plt.show()
```
