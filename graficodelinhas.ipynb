```python
# Bibliotecas necessarias

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variavel e funções

X = np.linspace(0,5,11)  # variavel

Y = np.power(X-2,2.0)    # Funções
Z = np.exp(X/2.0)
```

```python
 #Gráfico de Y em função de X

 plt.plot(X,Y)
plt.show()
```

```python
# Inserindo nome aos eixos e titulo

plt.plot(X,Y)
plt.xlabel('segundo eixo')
plt.ylabel('primeiro eixo')
plt.title('eixos')
plt.show()
```

```python
# Plotando dois graficos em uma só imagem e alterando suas cores

plt.plot(X,Y,color='black',label='Y(X)')
plt.plot(X,Z,color='pink',label='Z(X)')
plt.xlabel('primeiro eixo')
plt.ylabel('segundo eixo')
plt.title('eixos que se cruzam')
plt.legend()
plt.show()
```
