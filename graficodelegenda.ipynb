```python
# Criando um gráfico de legenda
```

```python
# Bibliotecas necessarias

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variavel e funções

X = np.linspace(0,5,11)  #variavel

Y = np.power(X-2,2.0)   #funções
Z = np.exp(X/2.0)
```

```python
# Adicionando as equações na legenda

plt.plot(X,Y,label='$Y(X)=(X-2)^2$', color = 'green')
plt.plot(X,Z,label='$Z(X)=e^{(x/2)}$', color = 'red')
plt.xlabel('Vertical')
plt.ylabel('Horizontal')
plt.title('crescimento')
plt.legend()
plt.show()
```

```python
# Criando novas variaveis

X = np.linspace(0,2.0*np.pi,361)
Y = np.cos(X/2)
Z = np.sin(X/4)
```

```python
plt.plot(X,Y,label='$\cos(X/2)$',color='black')
plt.plot(X,Z,label='$\sin(X/4)$',color='green',linestyle='-.')
plt.legend()
```
