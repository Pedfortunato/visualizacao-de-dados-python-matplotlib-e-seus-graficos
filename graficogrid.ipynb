```python
# bibliotecas necessarias

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variavel e funções

X = np.linspace(0,5,11)  # variavel
```

```python
Y = np.power(X-2,2.0)    # funções
Z = np.exp(X/2.0)
```

```python
# Criando grid

plt.plot(X, Y, label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.grid(True, linestyle='-', color='black', which='major')
plt.show()
```

```python
# Alterando a cor do fundo

plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.grid(linestyle = '--' , which = 'major' , color = 'grey')
plt.legend()
plt.show()
```
