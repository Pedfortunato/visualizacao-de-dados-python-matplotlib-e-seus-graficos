```python
# Bibliotecas necessarias

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando labels e valores

# Top5 artistas do spotify com mais ouvintes mensais  ! valores em milhões

labels = ["The Weeknd","Taylor Swift","Ed Sheeran","	Miley Cyrus","Shakira"]
valores = [106.96,85.36,84.09,80.77,76.90]
```

```python
# Criando lista com cores

cores = ['red','green','blue','orange','cyan']
```

```python
# Não podemos esquecer o raio do circulo

raio = 0.7
```

```python
# Plotando o gráfico

plt.pie(valores,labels=labels,colors=cores)
circulo = plt.Circle((0,0),raio,color='white')
p = plt.gcf()
p.gca().add_artist(circulo)
plt.show()
```

```python
# ADD porcentagem

plt.pie(valores,labels=labels,colors=cores,autopct='%1.1f%%')
circulo = plt.Circle((0,0),raio,color='white')
p = plt.gcf()
p.gca().add_artist(circulo)
plt.show()
```
