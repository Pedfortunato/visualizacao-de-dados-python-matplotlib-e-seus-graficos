```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Iremos fazer um levantamento de análise de desempenho dos jogadores do Vasco da Gama, nas duas primeiras rodadas do Brasileirão série A
```

```python
#criando labels
labels = ["Gabriel pec","Léo Jardim","Jair","Alex Teixeira","Marlon Gomes","Andrey santos","Lucas Piton","Figueiredo","Robson","Léo","Rodrigo","Luca Orellano","Capasso","Pedro Raul",""]
```

```python
#criando valores
valores = [9.0,7.75,7.5,6.5,5.0,7.0,7.8,6.0,6.5,7.75,6.65,6.60,6.0,6.9]
```

```python
# criando gráfico de barras
plt.bar(labels,valores)
```

```python
#inserindo nome aos eixos
plt.bar(labels,valores)
plt.xlabel('Nome atletas')
plt.ylabel('Média por jogo')
```

```python
# selecionando a cor
plt.bar(labels,valores,color='black')
plt.xlabel('Nome atletas')
plt.ylabel('Média por jogo')
```

```python
#modelo horizontal
plt.barh(labels,valores,color='red')
plt.xlabel('média por jogo')
plt.ylabel('Nome atletas')
```

```python
#colocando os Valores exatos acima
plt.bar(labels,valores,color='red')
plt.xlabel('média por jogo')
plt.ylabel('Nome atletas')
for x,y in zip(labels,valores):
    plt.text(x, y+0.05, '%d' % y, ha='center' , va= 'bottom')
```

```python
# Criando comparattivos, valores = minutos por partida
valores2 = [84,90,85,68,23,85,85,15,88,90,81,8,13,90]
```

```python
largura = 0.3
pos1 = np.arange(len(valores))
pos2 = [x + largura for x in pos1]

plt.bar(pos1,valores, color='red', width=largura, label='Média pontuação ')
plt.bar(pos2, valores2, color='black', width=largura, label='Média minutos')
plt.legend()
```

```python
plt.bar(pos1,valores,label = "Barra 1",color = 'black')
plt.bar(pos1,valores2,label = "Barra 2",bottom = valores)
```

```python
plt.barh(pos1,valores,label = "Barra 1")
plt.barh(pos1,valores2,label = "Barra 2",left = valores)
```
