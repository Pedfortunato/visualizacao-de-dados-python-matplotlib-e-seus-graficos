```python
# Bibliotecas necessarias

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variavel x com numeros aleatorios

X = np.random.rand(200)
```

```python
# plotando histograma

plt.hist(X)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.show()
```

```python
# As bins podem ser alteradas ou definidas a sua escola ex:

plt.hist(X,bins=30)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=70')
plt.show()
```

```python
plt.hist(X,bins=1)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=70')
plt.show()
```

```python
# Alterando as cores

plt.hist(X,bins=70,color='green')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=70')
plt.show()
```

```python
# Mais de um histograma em uma s imagem

X1 = np.random.normal(0,1,size = 800)
X2 = np.random.normal(3,1,size = 800)
X3 = np.random.normal(5,1,size = 800)

plt.hist(X1,bins=10,color='red')
plt.hist(X2,bins=10,color='blue')
plt.hist(X3,bins=10,color='green')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Multiplos histogramas')
plt.show()
```

```python
# Distribuio comulativa

X = np.random.rand(200)

plt.hist(X,bins=20,color='black', edgecolor='white',cumulative=True)
plt.xlabel('Crescimento')
plt.ylabel('Contagem')
plt.title('Histograma cumulativo')
plt.show()
```
